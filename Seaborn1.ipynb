```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

try:
    df = pd.read_csv('/content/FuelEfficiency(1).csv')
except FileNotFoundError:
    print("Please provide a valid path to your CSV file.")
    exit()  # or handle the error appropriately
print(df.head)

# 1. Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['# Gears'], df['CombMPG'], alpha=0.5)  # alpha for transparency
plt.xlabel('Number of Gears')
plt.ylabel('Combined MPG')
plt.title('Scatter Plot of Gears vs. MPG')
plt.show()

# 2. lmplot (Linear Regression Plot) # This is less code to do the same thing

sb.lmplot(x='# Gears', y='CombMPG', data=df)
plt.title('lmplot of Gears vs. MPG')
plt.show()

# 3. jointplot (Distribution and Correlation)
sb.jointplot(x='# Gears', y='CombMPG', data=df, kind='scatter') # You can change kind to 'kde' for density plots
plt.suptitle('jointplot of Gears vs. MPG', y=1.02) # Adjust title position
plt.show()

# 4. Box Plot (Distribution per Gear Category)
plt.figure(figsize=(8, 6))
sb.boxplot(x='# Gears', y='CombMPG', data=df)
plt.xlabel('Number of Gears')
plt.ylabel('Combined MPG')
plt.title('Box Plot of MPG by Number of Gears')
plt.show()

# 5. Swarm Plot (Distribution per Gear Category - detailed)
plt.figure(figsize=(10, 6)) # Adjust figure size as needed
sb.swarmplot(x='# Gears', y='CombMPG', data=df, size=5) # Adjust size as needed
plt.xlabel('Number of Gears')
plt.ylabel('Combined MPG')
plt.title('Swarm Plot of MPG by Number of Gears')
plt.show()
```
