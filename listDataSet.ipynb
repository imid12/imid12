```python
"""
This program manages a list of integers. It allows the user to add,
change, insert, get, print, total, average, remove, and clear values
within the list through a menu-driven interface.
"""

class dataSetIntegerList:
   # A class to manipulate a list of integers.

    def __init__(self):
        # Initializes the list.
        self.integer_list = []

    def addValues(self):
        # Adds values to the list until 0 is entered.
        print("Enter values to add to the list (enter 0 to finish):")
        while True:
            try:
                value = int(input("Enter an integer: "))
                if value == 0:
                    break
                self.integer_list.append(value)
            except ValueError:
                print("Invalid input. Please enter an integer.")

    def changeValue(self):
        # Changes the value at a given location.
        if not self.integer_list:
            print("The list is empty. Cannot change a value.")
            return
        try:
            index = int(input("Enter the index of the value to change: "))
            if 0 <= index < len(self.integer_list):
                new_value = int(input("Enter the new value: "))
                self.integer_list[index] = new_value
                print("Value changed successfully.")
            else:
                print("Error: Invalid index.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

    def insertValue(self):
        # Inserts a value at a given location.
        try:
            index = int(input("Enter the index to insert the value: "))
            new_value = int(input("Enter the value to insert: "))
            if 0 <= index <= len(self.integer_list):
                self.integer_list.insert(index, new_value)
                print("Value inserted successfully.")
            else:
                print("Error: Invalid index.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

    def getIndex(self):
        # Prints the value at a given location.
        if not self.integer_list:
            print("The list is empty. Cannot get a value.")
            return
        try:
            index = int(input("Enter the index of the value to get: "))
            if 0 <= index < len(self.integer_list):
                print(f"The value at index {index} is {self.integer_list[index]}")
            else:
                print("Error: Invalid index.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

    def getSize(self):
        # Prints the number of elements in the list.
        print(f"The size of the list is {len(self.integer_list)}")

    def printList(self):
        # Displays all the values in the list.
        if not self.integer_list:
            print("The list is empty.")
        else:
            print("The values in the list are:", self.integer_list)

    def calculateTotal(self):
        # Prints the total of all the values in the list.
        if not self.integer_list:
            print("The list is empty. Total is 0.")
            return 0
        total = sum(self.integer_list)
        print("The total of the values in the list is:", total)
        return total

    def calculateAverage(self):
        # Prints the average of all the values in the list.
        if not self.integer_list:
            print("The list is empty. Average is 0.")
            return 0
        total = self.calculate_total()  # Reuse the calculate_total function
        average = total / len(self.integer_list)
        print("The average of the values in the list is:", average)
        return average

    def removeValue(self):
        # Attempts to remove a value.
        if not self.integer_list:
            print("The list is empty. Cannot remove a value.")
            return
        try:
            value_to_remove = int(input("Enter the value to remove: "))
            try:
                self.integer_list.remove(value_to_remove)
                print("Remove successful.")
            except ValueError:
                print("Value not found.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

    def removeLocation(self):
        # Removes the value at a given location.
        if not self.integer_list:
            print("The list is empty. Cannot remove a value.")
            return
        try:
            index = int(input("Enter the index of the value to remove: "))
            if 0 <= index < len(self.integer_list):
                removed_value = self.integer_list.pop(index)
                print(f"The value {removed_value} was removed from index {index}.")
            else:
                print("Error: Invalid index.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

    def clearList(self):
        # Empties the list.
        self.integer_list.clear()
        print("The list has been cleared.")

def main():
    # This the main function used to set up the menu and call the functions.
    listDataSet = dataSetIntegerList()

    while True:
        print("\nMenu:")
        print("1. Add values")
        print("2. Change a value")
        print("3. Insert a value")
        print("4. Get Index")
        print("5. Get Size")
        print("6. Print")
        print("7. Total")
        print("8. Average")
        print("9. Remove a value")
        print("10. Remove a location")
        print("11. Clear")
        print("12. Exit")

        try:
            choice = int(input("Enter your choice: "))
        except ValueError:
            print("Invalid choice. Please enter an integer between 1 and 12.")
            continue

        match choice:
          case 1: listDataSet.addValues()
          case 2: listDataSet.changeValue()
          case 3: listDataSet.insertValue()
          case 4: listDataSet.getIndex()
          case 5: listDataSet.getSize()
          case 6: listDataSet.printList()
          case 7: listDataSet.calculateTotal()
          case 8: listDataSet.calculateAverage()
          case 9: listDataSet.removeValue()
          case 10: listDataSet.removeLocation()
          case 11: listDataSet.clearList()
          case 12:
                   print("Exiting program.")
                   break
          case _: print("Invalid choice. Please enter an integer between 1 and 12.")

if __name__ == "__main__":
    main()
```
